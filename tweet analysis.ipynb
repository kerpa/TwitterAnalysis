```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np # linear algebra
import advertools as at # package for tweet analysis
```

```python
df = pd.read_excel (r'C:\Users\jeanc\Desktop\netflix.xlsx', sheet_name='tweets') # Read the data from the file
df.head() # Print the first 5 rows of the dataframe
```

```python
# Create a new column with the length of the tweet
df['Length of tweet'] = df['Text'].apply(len)

#Create a new column with the number of words in the tweet
df['Number of words'] = df['Text'].apply(lambda x: len(x.split()))
```

```python
# Create a new column if there is a link in the tweet
df['Link'] = df['Text'].apply(lambda x: 1 if 'http' in x else 0)

#create a new column with the kind of media in the tweet
a = (df['Media Type'] == 'photo')
b = (df['Media Type'] == 'video')
c = (df['Media Type'] == 'animated_gif')

df['Media'] = np.where(a, 1, np.where(b, 2, np.where(c, 3, 0)))
#df['Media'] = np.where(a, 1, np.where(b, 2, np.where(np.where(c, 3, 0), 0))) # If the media type is photo, video or animated gif, the value is 1, 2 or 3. If not, the value is 0.



""" for media in df['Media Type']:
    if media == 'photo':
        df.loc[:,(media,'Media')] = 1
    elif media == 'video':
        df.loc[:,(media,'Media')] = 2
    elif media == 'animated_gif':
        df.loc[:,(media,'Media')] = 3
    elif media == 'none':
        df.loc[:,(media,'Media')] = 0 """

#df['Media'] = df['Media Type'].apply(lambda x: 1 if 'photo' in x else (2 if 'video' in x else (3 if 'animated_gif' in x else 0)))
```

```python
df.head() # Print the first 5 rows of the dataframe
```

```python
#Create a new column if there is a hashtag in the tweet
df['hashtag'] = df['Text'].apply(lambda x: 1 if '#' in x else 0)

#Create a new column if there is a retweet in the tweet
df['retweet'] = df['Text'].apply(lambda x: 1 if 'RT' in x else 0)

#Create a new column if there is a mention in the tweet
df['mention'] = df['Text'].apply(lambda x: 1 if '@' in x else 0)
```

```python
#create a new column if there is an exclamation mark in the tweet
df['exclamation'] = df['Text'].apply(lambda x: 1 if '!' in x else 0)

#create a new column if there is a question mark in the tweet
df['question'] = df['Text'].apply(lambda x: 1 if '?' in x else 0)

#create a new column if there an emoji in the tweet
emoji_summary = at.extract_emoji(df['Text'])
df['emoji'] = emoji_summary['emoji_counts']
df['emoji'] = df['emoji'].apply(lambda x: 1 if x > 0 else 0)
```
